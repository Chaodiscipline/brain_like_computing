# Hopfield

```python
class Hopfield():
    def __init__(self, memory,row, col):
        # memory是待记忆的一组模式，用二维列表表示
        self.memory = memory
        self.row = row
        self.col = col
        self.N = len(memory[0])
        self.W = np.zeros((self.N, self.N), dtype = np.float32)
        
    def train(self):
        # 重置W
        self.W = np.zeros((self.N, self.N), dtype = np.float32)
        numArray = np.asarray(self.memory,dtype = np.int8)
        print(str(len(numArray))+" patterns")
        # 更新W
        for i in range(len(numArray)):
            v = numArray[i].reshape((1,self.N))
            self.W += np.dot(v.T,v)/self.N
        self.W[range(self.N),range(self.N)] = 0
        print('weights：\n',self.W)
        
    def predict(self, preArray):
        lastState = np.asarray(preArray,dtype = np.int8)
        for i in range(100):
            state = np.dot(self.W, lastState)
            state[state>=0] = 1
            state[state<0] = -1
            if (state==lastState).all():
                break
            else:
                lastState = state

        self.printNum(lastState)
#         print("迭代",i,"次收敛")
        return lastState
    
    def printNum(self, Num):
        a = np.asarray(Num)
        b = ['* ' if i==1 else '  ' for i in a]
        for i in range(self.row):
            print(''.join(b[:self.col]))
            b = b[self.col:]
```

# 加入噪声

```python
def addNoise(noiseType, originNum, noiseSize):
    noiseNum = np.asarray(originNum)
    if noiseType=='random':
        noisePosition = np.random.randint(30, size=noiseSize)
        for i in noisePosition:
            if noiseNum[i]==-1:
                noiseNum[i] = 1
            else:
                noiseNum[i] = -1
    elif noiseType=='half':
        for i in range(15,30):
            noiseNum[i] = -1
    else:
        print('no such noiseType.')
    return noiseNum
```

# 测试

```python
noiseType = 'half'
noiseSize = 2
toPredictNum = 'one'
print('toPredictNum = ',toPredictNum ,'  noiseType = ',noiseType,'  noiseSize = ',noiseSize)
print("原始数字:")
Hop.printNum(numDict[toPredictNum])
toPredict = addNoise( noiseType, numDict[toPredictNum],noiseSize)
print("加入噪声:")
Hop.printNum(toPredict)
print("联想结果:")
Hop.predict(toPredict)
```
